```python
import sys
sys.path.append('../src')
from data_manager import *

import os
import matplotlib.pyplot as plt

root_path = '../'
```

```python
jaffe_data_path = os.path.join(root_path, 'data/JAFFE')
jaffe_data_path
```

```python
image_files = list(get_image_paths(jaffe_data_path))

images = get_images(image_files)
labels = get_labels(image_files, 'jaffe')

print(f"Dataset length: {len(image_files)}")
print(f"Image shape: {images[0].shape}")
```

```python
plt.figure(figsize=(15, 3))

label_history = []
for i in range(len(images)):

    if labels[i] in label_history:
        continue
    label_history.append(labels[i])

    plt.subplot(1, 8, len(label_history))
    plt.imshow(images[i], cmap='gray')
    plt.title(labels[i])
    plt.axis('off')

plt.savefig('figures/jaffe.png')
plt.show()
```
